```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from matplotlib.colors import LinearSegmentedColormap
import ipywidgets as widgets
from IPython.display import display, clear_output
from ipywidgets import interact, FloatSlider

data = {
    'Weapon': ['Concussion Missile Launchers', 'Heavy Laser', 'Ion Batteries', 'Ion Cannons', 'Proton Torpedo Launchers', 'Quad Lasers', 'Torpedo Launchers', 'Turbolaser Battery', 'Turbolasers'],
    'Damage Type': ['Concussive (H)', 'Energy (H)', 'Ionic (H)', 'Ionic (H)', 'Explosive (H)', 'Energy (H)', 'Explosive (P)', 'Turbolaser (H)', 'Turbolaser (H)'],
    'Firepower': [40, 10, 60, 20, 60, 8, 1, 60, 25],
    'Maximum Hits': [3, 3, 3, 2, 2, 10, 3, 3, 3],
    'Minimum Damage': [40, 10, 60, 20, 60, 8, 100, 480, 60],
    'Maximum Damage': [40, 10, 60, 20, 60, 8, 150, 480, 60],
    'Optimum Range': [1, 1, 2, 1, 2, 1, 2, 2, 2],
    'Drop Off': [4.5, 4.5, 4.5, 4.5, 4.5, 4.5, 4.5, 2, 5],
    'Tracking': [6, 4, 1, 3, 2, 6, 1, 1, 1]
}

mounted_weap = pd.DataFrame(data)
mounted_weap

modifiers = {
    "Energy (H)": {
        "Fighters (A)": {"Fighters (D)": 2.0, "Gunboats (D)": 2.0, "Light Freighters (D)": 1.2, "Heavy Freighters (D)": 1.1, "Corvettes (D)": 0.4, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Gunboats (A)": {"Fighters (D)": 2.0, "Gunboats (D)": 2.0, "Light Freighters (D)": 1.8, "Heavy Freighters (D)": 1.0, "Corvettes (D)": 0.4, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Light Freighters (A)": {"Fighters (D)": 1.8, "Gunboats (D)": 1.0, "Light Freighters (D)": 1.0, "Heavy Freighters (D)": 1.0, "Corvettes (D)": 1.0, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Heavy Freighters (A)": {"Fighters (D)": 1.5, "Gunboats (D)": 1.0, "Light Freighters (D)": 1.0, "Heavy Freighters (D)": 1.0, "Corvettes (D)": 1.0, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Corvettes (A)": {"Fighters (D)": 4.0, "Gunboats (D)": 4.0, "Light Freighters (D)": 2.0, "Heavy Freighters (D)": 3.0, "Corvettes (D)": 2.0, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
    },
    "Explosive (H)": {
        "Fighters (A)": {"Fighters (D)": 0.3, "Gunboats (D)": 0.5, "Light Freighters (D)": 0.5, "Heavy Freighters (D)": 0.8, "Corvettes (D)": 1.0, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Gunboats (A)": {"Fighters (D)": 0.5, "Gunboats (D)": 0.5, "Light Freighters (D)": 1.0, "Heavy Freighters (D)": 1.0, "Corvettes (D)": 0.7, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Light Freighters (A)": {"Fighters (D)": 0.8, "Gunboats (D)": 1.0, "Light Freighters (D)": 1.0, "Heavy Freighters (D)": 1.0, "Corvettes (D)": 1.0, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Heavy Freighters (A)": {"Fighters (D)": 1.0, "Gunboats (D)": 1.0, "Light Freighters (D)": 1.0, "Heavy Freighters (D)": 1.0, "Corvettes (D)": 1.0, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Corvettes (A)": {"Fighters (D)": 2.0, "Gunboats (D)": 3.0, "Light Freighters (D)": 1.0, "Heavy Freighters (D)": 3.0, "Corvettes (D)": 2.0, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
    },
    "Ionic (H)": {
        "Fighters (A)": {"Fighters (D)": 0.4, "Gunboats (D)": 0.4, "Light Freighters (D)": 0.8, "Heavy Freighters (D)": 0.7, "Corvettes (D)": 1.0, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Gunboats (A)": {"Fighters (D)": 0.4, "Gunboats (D)": 0.4, "Light Freighters (D)": 1.0, "Heavy Freighters (D)": 1.0, "Corvettes (D)": 1.0, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Light Freighters (A)": {"Fighters (D)": 1.0, "Gunboats (D)": 1.0, "Light Freighters (D)": 1.0, "Heavy Freighters (D)": 1.0, "Corvettes (D)": 1.0, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Heavy Freighters (A)": {"Fighters (D)": 1.5, "Gunboats (D)": 1.0, "Light Freighters (D)": 1.0, "Heavy Freighters (D)": 1.0, "Corvettes (D)": 1.0, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Corvettes (A)": {"Fighters (D)": 3.0, "Gunboats (D)": 3.5, "Light Freighters (D)": 2.5, "Heavy Freighters (D)": 3.0, "Corvettes (D)": 3.0, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
    },
    "Concussive (H)": {
        "Fighters (A)": {"Fighters (D)": 0.5, "Gunboats (D)": 0.5, "Light Freighters (D)": 0.5, "Heavy Freighters (D)": 0.6, "Corvettes (D)": 0.7, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Gunboats (A)": {"Fighters (D)": 0.5, "Gunboats (D)": 0.5, "Light Freighters (D)": 0.5, "Heavy Freighters (D)": 0.6, "Corvettes (D)": 0.7, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Light Freighters (A)": {"Fighters (D)": 0.5, "Gunboats (D)": 0.5, "Light Freighters (D)": 0.5, "Heavy Freighters (D)": 0.6, "Corvettes (D)": 0.7, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Heavy Freighters (A)": {"Fighters (D)": 0.5, "Gunboats (D)": 0.5, "Light Freighters (D)": 0.5, "Heavy Freighters (D)": 0.6, "Corvettes (D)": 0.7, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
        "Corvettes (A)": {"Fighters (D)": 0.5, "Gunboats (D)": 0.5, "Light Freighters (D)": 0.5, "Heavy Freighters (D)": 0.6, "Corvettes (D)": 0.7, "Satellites (D)": 1.0, "Cargo Containers (D)": 1.0},
    },
    "Turbolaser (H)": {
    "Fighters (A)": {"Fighters (D)": 1.5, "Gunboats (D)": 1.4, "Light Freighters (D)": 1.3, "Heavy Freighters (D)": 1.1, "Corvettes (D)": 0.8},
    "Gunboats (A)": {"Fighters (D)": 1.5, "Gunboats (D)": 1.4, "Light Freighters (D)": 1.0, "Heavy Freighters (D)": 1.0, "Corvettes (D)": 0.6},
    "Light Freighters (A)": {"Fighters (D)": 1.2, "Gunboats (D)": 1.0, "Light Freighters (D)": 1.0, "Heavy Freighters (D)": 1.0, "Corvettes (D)": 0.9},
    "Heavy Freighters (A)": {"Fighters (D)": 1.0, "Gunboats (D)": 1.0, "Light Freighters (D)": 1.0, "Heavy Freighters (D)": 1.0, "Corvettes (D)": 0.7},
    "Corvettes (A)": {"Fighters (D)": 2.0, "Gunboats (D)": 2.0, "Light Freighters (D)": 1.5, "Heavy Freighters (D)": 1.4, "Corvettes (D)": 1.2},
    }
}

ship_class = pd.DataFrame(modifiers)

data = {
    "Length Ratio": ["> 200", "> 150", "> 100", "> 75", "> 50", "> 25", "> 10", "> 5", "> 2", "1:1", "< 2", "< 5", "< 10", "< 25", "< 50", "< 75", "< 100", "< 150", "< 200"],
    "Modifier (%)": [300, 200, 150, 130, 125, 120, 115, 110, 105, 100, 95, 90, 85, 80, 75, 70, 65, 60, 55]
}

length_modifier_df = pd.DataFrame(data)

hit_chances_data = {
    1: [0.30, 0.35, 0.40, 0.45, 0.50, 0.55, 0.60, 0.65, 0.70, 0.75],
    2: [0.25, 0.30, 0.35, 0.40, 0.45, 0.50, 0.55, 0.60, 0.65, 0.70],
    3: [0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50, 0.55, 0.60, 0.65],
    4: [0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50, 0.55, 0.60],
    5: [0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50, 0.55],
    6: [0.07, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50],
    7: [0.04, 0.07, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.45],
    8: [0.03, 0.04, 0.07, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40],
    9: [0.02, 0.03, 0.04, 0.07, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35],
    10: [0.01, 0.02, 0.03, 0.04, 0.07, 0.10, 0.15, 0.20, 0.25, 0.30]
}

damage_type_multipliers_data = {
    'Damage Type': ['Energy (H)', 'Explosive (H)', 'Ionic (H)', 'Concussive (H)', 'Turbolaser (H)'],
    'Multiplier': [1.00, 0.50, 1.00, 1.00, 0.50]
}

damage_type_multipliers_df = pd.DataFrame(damage_type_multipliers_data)

hit_chances = pd.DataFrame(hit_chances_data)
hit_chances.columns = range(1, 11)

# Define the tracking skill equation
def calculate_actual_tracking(base_weapon_tracking, combat_skill):
    return base_weapon_tracking + (combat_skill / 2)

# Define the maneuverability equation for the defender
def calculate_actual_maneuverability(ship_maneuverability, pilot_skill):
    required_piloting_skill = max(0, ship_maneuverability - 5)
    return ship_maneuverability + ((pilot_skill - required_piloting_skill) / 2)

def interpolate(value, x0, x1, y0, y1):
    """Linear interpolation."""
    return y0 + (y1 - y0) * (value - x0) / (x1 - x0)

def lookup_hit_chance(actual_maneuver, actual_tracking):
    # Determine the rows and columns to use for interpolation
    row_floor, row_ceil = int(math.floor(actual_maneuver)), int(math.ceil(actual_maneuver))
    col_floor, col_ceil = int(math.floor(actual_tracking)), int(math.ceil(actual_tracking))
    
    # Ensure we're not going out of bounds
    max_row_index = hit_chances.index.max()  # Maximum row index
    max_col_label = hit_chances.columns.max()  # Maximum column label

    row_ceil = min(row_ceil, max_row_index)
    col_ceil = min(col_ceil, max_col_label)
    
    # If both values are whole numbers
    if row_floor == row_ceil and col_floor == col_ceil:
        return hit_chances.loc[row_floor, col_floor]
    
    # If only the maneuverability is a non-whole number
    if row_floor != row_ceil and col_floor == col_ceil:
        return interpolate(actual_maneuver, row_floor, row_ceil, hit_chances.loc[row_floor, col_floor], hit_chances.loc[row_ceil, col_floor])
    
    # If only the tracking is a non-whole number
    if row_floor == row_ceil and col_floor != col_ceil:
        return interpolate(actual_tracking, col_floor, col_ceil, hit_chances.loc[row_floor, col_floor], hit_chances.loc[row_floor, col_ceil])
    
    # If both values are non-whole numbers, bilinear interpolation
    value1 = interpolate(actual_maneuver, row_floor, row_ceil, hit_chances.loc[row_floor, col_floor], hit_chances.loc[row_ceil, col_floor])
    value2 = interpolate(actual_maneuver, row_floor, row_ceil, hit_chances.loc[row_floor, col_ceil], hit_chances.loc[row_ceil, col_ceil])
    return interpolate(actual_tracking, col_floor, col_ceil, value1, value2)

# Define the range modifier equation
def calculate_range_modifier(optimum_range, current_range, weapon_drop_off):
    delta_range = abs(optimum_range - current_range)
    return max(0, (1 - (0.95 / (1 + math.exp(-2 * (delta_range - weapon_drop_off)))) - (0.003 * delta_range)))

# Define the length modifier equation
def calculate_length_modifier(length_modifier_key):
    return length_modifier_df.loc[length_modifier_df["Length Ratio"] == length_modifier_key, "Modifier (%)"].values[0] / 100

# Define the maneuver comparison modifier
def calculate_maneuver_comparison(attacker_maneuver, defender_maneuver):
    delta_maneuver = attacker_maneuver - defender_maneuver
    return 1 + (delta_maneuver / max(attacker_maneuver, defender_maneuver))

# Define the damage equation with armor reduction
def calculate_damage(din, firepower, armor):
    return din * (firepower / (firepower + armor))

def compute_hit_and_damage(combat_skill, ship_maneuverability, pilot_skill, length_modifier_key, attacker_maneuver, defender_maneuver, current_range, armor, base_weapon_tracking, optimum_range, weapon_drop_off, max_hits, min_damage, max_damage, damage_modifier, weapon_name): 
    actual_maneuver = calculate_actual_maneuverability(ship_maneuverability, pilot_skill)
    
    # Calculate actual_tracking here
    actual_tracking = calculate_actual_tracking(base_weapon_tracking, combat_skill)
    
    hit_chance = lookup_hit_chance(actual_maneuver, actual_tracking)
    range_modifier = calculate_range_modifier(optimum_range, current_range, weapon_drop_off)
    length_modifier = calculate_length_modifier(length_modifier_key)
    maneuver_comparison = calculate_maneuver_comparison(attacker_maneuver, defender_maneuver)
    final_hit_chance = hit_chance * range_modifier * length_modifier * maneuver_comparison
    
    din = (min_damage + max_damage) / 2  
    firepower = din * max_hits
    
    # Fetch the damage type multiplier based on the weapon name
    damage_type = mounted_weap[mounted_weap['Weapon'] == weapon_name]['Damage Type'].values[0]
    damage_type_multiplier = damage_type_multipliers_df[damage_type_multipliers_df['Damage Type'] == damage_type]['Multiplier'].values[0]

    final_damage = final_hit_chance * (calculate_damage(din, firepower, armor) * damage_modifier * damage_type_multiplier)
    return final_hit_chance, final_damage
    
ranges = list(range(0, 9))
armors = list(range(20, 76))

def compute_damage_map(weapon_name, combat_skill, ship_maneuverability, pilot_skill, length_modifier_key, attacker_maneuver, damage_modifier):

    # Fetch weapon details from the DataFrame
    weapon_details = mounted_weap[mounted_weap['Weapon'] == weapon_name].iloc[0]

    base_weapon_tracking = weapon_details['Tracking']
    optimum_range = weapon_details['Optimum Range']
    weapon_drop_off = weapon_details['Drop Off']
    max_hits = weapon_details['Maximum Hits']
    min_damage = weapon_details['Minimum Damage']
    max_damage = weapon_details['Maximum Damage']

    # Initialize damage_values array
    damage_values = np.zeros((len(armors), len(ranges)))  # Initialize with zeros
    
    for i, current_range in enumerate(ranges):
        for j, armor in enumerate(armors):
            _, damage = compute_hit_and_damage(combat_skill, ship_maneuverability, pilot_skill, length_modifier_key, attacker_maneuver, ship_maneuverability, current_range, armor, base_weapon_tracking, optimum_range, weapon_drop_off, max_hits, min_damage, max_damage, damage_modifier, weapon_name)
            damage_values[j][i] = damage
    return damage_values

# Create sliders for combat_skill, pilot_skill, ship_maneuverability, and attacker_maneuver
combat_skill_slider = widgets.IntSlider(value=3, min=0, max=5, step=1, description='Combat Skill (A):')
pilot_skill_slider = widgets.IntSlider(value=3, min=0, max=5, step=1, description='Pilot Skill (D):')
ship_maneuverability_slider = widgets.IntSlider(value=6, min=1, max=10, step=1, description='Defender Maneuver:')
attacker_maneuver_slider = widgets.IntSlider(value=7, min=1, max=10, step=1, description='Attacker Maneuver:')

# Create a slider for length_modifier_key using the "Length Ratio" values
length_modifier_slider = widgets.Dropdown(
    options=data["Length Ratio"],
    value=data["Length Ratio"][9],  # Default to "1:1"
    description='Length Ratio:'
)

weapon_dropdown = widgets.Dropdown(
    options=mounted_weap['Weapon'].tolist(),
    value=mounted_weap['Weapon'].iloc[0],  # Default to the first weapon
    description='Weapon:'
)

attacker_dropdown = widgets.Dropdown(
    options=["Fighters (A)", "Gunboats (A)", "Light Freighters (A)", "Heavy Freighters (A)", "Corvettes (A)"],
    value="Fighters (A)",  # Default to "Fighters (A)"
    description='Attacker:'
)

defender_dropdown = widgets.Dropdown(
    options=["Fighters (D)", "Gunboats (D)", "Light Freighters (D)", "Heavy Freighters (D)", "Corvettes (D)", "Satellites (D)", "Cargo Containers (D)"],
    value="Fighters (D)",  # Default to "Fighters (D)"
    description='Defender:'
)

def generate_heatmap(weapon_name, combat_skill, pilot_skill, ship_maneuverability, attacker_maneuver, length_ratio, attacker_type, defender_type):
    length_modifier_key = length_ratio
    
    # Look up the 'Damage Type' based on the selected weapon
    damage_type = mounted_weap[mounted_weap['Weapon'] == weapon_name]['Damage Type'].values[0]
    
    # Fetch the modifier from the 'modifiers' dictionary
    damage_modifier = modifiers[damage_type][attacker_type][defender_type]
    
    damage_values = compute_damage_map(weapon_name, combat_skill, ship_maneuverability, pilot_skill, length_modifier_key, attacker_maneuver, damage_modifier)
    
    # Define the custom colormap
    colors = [(0, "black"), (0.125, "red"), (0.25, "orange"), (0.375, "yellow"), (0.5, "green"), (0.625, "blue"), (0.75, "indigo"), (0.875, "violet"), (1, "white")]
    quad_cmap = LinearSegmentedColormap.from_list("quad", colors)

    fig, ax = plt.subplots(figsize=(15, 6))  # Made it wider
    cax = ax.imshow(damage_values, cmap=quad_cmap, origin="lower", vmin=0, vmax=np.max(damage_values))
    plt.colorbar(cax, ax=ax, label="Final Damage")

    ax.set_title(f'Heatmap of Final Damage for {weapon_name}')
    ax.set_xlabel('Range')
    ax.set_ylabel('Armor')

    # Adjusting tick marks for better visuals
    tick_spacing_ranges = 2
    ax.set_xticks(np.arange(0, len(ranges), tick_spacing_ranges))
    ax.set_xticklabels([str(ranges[i]) for i in range(0, len(ranges), tick_spacing_ranges)])

    tick_spacing_armors = 5  # adjusting this for better visuals
    ax.set_yticks(np.arange(0, len(armors), tick_spacing_armors))
    ax.set_yticklabels([str(armors[i]) for i in range(0, len(armors), tick_spacing_armors)])

    plt.show()

interact(generate_heatmap,
         weapon_name=weapon_dropdown,
         combat_skill=combat_skill_slider,
         pilot_skill=pilot_skill_slider,
         ship_maneuverability=ship_maneuverability_slider,
         attacker_maneuver=attacker_maneuver_slider,
         length_ratio=length_modifier_slider,
         attacker_type=attacker_dropdown,
         defender_type=defender_dropdown)
```
